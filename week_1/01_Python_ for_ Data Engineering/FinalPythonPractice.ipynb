```python
# Problem 1: Program that takes a number as input
# and return wheather number is positive, negative or zero

number = int(input("Please enter a number of your choice"))

def numberChecker(num):
    if number  == 0:
        return f'The number: {num} is a zero'
    elif number > 0:
        return f'The number: {num} is positive'
    else:
        return f'The number: {num} is negative'

FinalOutput = numberChecker(number)
print(FinalOutput);
```

```python
# Problem 2: From the give list of numbers identify the even or Odd 
number = [12, 7, 9, 20, 33, 40]
evenNumbers = []
oddNumbers = []

def EvenAndOddFilters(number):
    for num in number:
        if num % 2 == 0:
            evenNumbers.append(num)
            print(f"The even numbers are: {evenNumbers}")
        if num % 2 != 0:
            oddNumbers.append(num)
            print(f"The odd numbers are: {oddNumbers}")
            
EvenAndOddFilters(number)
```

```python
# Problem 3: Sum of Elements in a List
listItems = [1, 3, 4, 12, 56, 78, 99, 100]

def additionOfList(items):
    total = 0
    for item in items:
        total += item 
    return f'The total sum of the list items is: {total}'

finalOutput = additionOfList(listItems)
print(finalOutput)
        
```

```python
# Problem 5: Count Vowels in a String

string = input("Enter a string value")
vowelsChar = 'aeiouAEIOU'

def stringCounter(string):
    count = 0
    for char in string:
        if char in vowelsChar:
            count += 1
    return count;

output = stringCounter(string)

print(output)

```

```python
# Problem 6: Remvoing the duplicates from a list

listItems = ['apple', 'banana', 'mango', 'banana', 'orange', 'grapes', 'kiwi', 'apple']

uniqueList = []

for items in listItems:
    if items not in uniqueList:
        uniqueList.append(items)
        
print(uniqueList)
```
