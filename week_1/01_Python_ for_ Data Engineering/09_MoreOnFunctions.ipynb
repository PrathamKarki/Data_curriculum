```python
# FIRST CLASS FUNCTION

# Functions assigned to another variables

def greeting():
    print("hello")

greet_something = greeting # assigning function as a variable

greet_something()
```

```python
# Function can be passed as arguments
def greet(): # here greet is function pass as argument as call_function(greet)
    print("hello")

def call_function(func):
    func()

call_function(greet)
```

```python
# FUNCTION can retrun other function 
def outer():
    def inner():
        print("this is inner function")
    return inner

func = outer()
func()
```
