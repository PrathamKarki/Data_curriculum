```python
# Creating of class and objects in python

class students:
    x = 'hello'

# using created class 'students'
obj1 = students()
print(obj1.x)

# Creating of more classes
class people:
    x = 'ram'
    y= 'hari'

obj2 = people()
print(obj2.x)
print(obj2.y)


class Student:
    def __init__(self, name, age, rollNo, address):
        self.name = name
        self.age = age
        self.rollNo = rollNo
        self.address = address

s1 = Student('goofy', 19, 2040, 'kathmandu')
print(s1.name)
```

```python
# Class and method , objects

class Rectangle:
    def __init__(self, length, breadth):
        self.length = length
        self.breadth = breadth

    def areaOfRectangle(self):
        print (f'Area of rectangle is {self.length * self.breadth} sq.')

obj = Rectangle(14, 10)
obj.areaOfRectangle()
```
