```python
# Checking for multiple variables 
a = 10
b = 10.0

if isinstance(a, int) and isinstance(b, float):
    print("Both a  and b  are valid types")
else:
    print("Both a and b are invalid types")

```

```python
# String formatting 
name = "Pratham"
age = 22
description = "My name is " + name + " and i am " + str(age) + " years old."
print(description)


# using the f string method

descriptin2 = f"my name is {name} and i am {age} years old."
print(descriptin2)
```
