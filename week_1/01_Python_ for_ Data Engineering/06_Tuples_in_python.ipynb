```python
# Tuples in Python 
# Tuples are exactly similar to list but uses the parenthesis instead of the square brackket

a = (1,2,5,12,18)

# we can write the syntax even as 
a = 1, 3, 5, 12, 18
# hence it is equivalent to the above syntax

print(a)

#checking the type of the tuple

type(a)
```

```python
# Tuples are immutables that means they cannot be modified unlike of list

b = (1, 2, 15, 17, 34)

# b[0] = 15
```

```python
# Looping through tuples

foods = ['Pizza', 'Burger', 'rice', 'eggs', 'bacon']

for food in foods:
    print(food, end = " ")
```

```python
# method in tuples

foods = ('Pizza', 'Burger', 'rice', 'eggs', 'bacon')

print(foods.count('Burger'))
print(foods.index('bacon'))

# Unpacking of tuples
a, b, c =  (10, 20, 30)
print(a)
print(b)
print(c)
```

```python
# Some practice problems in Tuples

numbers = (1, 4, 13, 45, 56)
print(numbers[0])
print(numbers[-1])

# Creating a tuple of 3 cities and printing each one using a loop
cities = ('Kathmandu', 'Pokhara', 'Biratnagar')

for city in cities:
    print(city)

# Counting how many times 2 appear
nums = (1, 2, 2, 3, 2, 4)
count = 0;

for num in nums:
    if num == 2:
        count += 1
print(count)

# Find the index of "banana"
fruits = 'apple', 'banana', 'mango'

fruits.index('banana')

# Unpaching of the tuples in python
name, age, rollNo  = ('pratham', 24, 7030)
print(name, age, rollNo)

# Converting of list into tuples
planets = ['Mercury', 'Venus', 'Earth', 'Mars', 'Jupiter']

conversionTuples = tuple(planets)
print(conversionTuples)

# Checking if the element exist in tupple or not
nums = (10, 20, 30, 40)

25 in nums

# why this gives error
t = (1, 2, 3)
t[1] = 100

# because tuples are immutables unlike list, the modifying of values in tuples doesn't work 
```

```python
# Converting the tuples into list and then mutation the value and converting back to the tupl
data = (10, 20, 30, 40)

data = list(data)
print(type(data))
data[1] = 25

data = tuple(data)
print(type(data))
print(data)

# Print only the even numbers from the tuples
t = (1, 2, 3, 4, 5)

for even in t:
    if even % 2 == 0:
        print(even)
```
