```python
# Arithmetic operators in python 
# arithmetic operators are those basic operators in python used to perform calculation such as addition, subtraction, multiplication, divisiono

num1 = 12
num2 = 78

# addition 
addition = num1 + num2  

# subtraction 
sub = num2 - num1

# multiplication 
mul = num1 * num2

# diviison 
div = num2/ num1

# mixed with brackets
mixed = ((num1 * num2)/ num2) * num1

print(addition)
print(sub)
print(mul)
print(div)
print(mixed)
```

```python
# modulo operator: returns the reminder based on the division performed
num1 = 112
num2 = 40

modulsOp = num1 % num2
print(modulsOp)

# Exponential Operator
num1 = 5
num2 = 2
power = 5 ** 2
print(power)

# multiplying of the string 
str = "hello " * 5 
print(str)
```

```python
# Logical Operators and Conditional in pYTHON
# Checking for the even or odd
number = 13;
if number % 2 == 0:
    print("The given number is even:", number)
else:
    print("The given number is odd:", number)

if number >=1 and number <= 100:
    print("The given number is between 1 and 100")
else:
    print("The given number is not between 1 and 100")
```
