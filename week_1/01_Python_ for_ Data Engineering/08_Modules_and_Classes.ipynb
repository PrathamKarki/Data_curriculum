```python
# Classes in Python 
# Python is object oriended programming langauge, where the concept of classes
# in python acts as a blueprint for creating objects

# Creating class name "myClass"
class myClass:
    x = 56
    y= 98

# using the created class 'myClass" to create object
obj1 = myClass()
print(obj1.y)

#deleting of the obj 
del obj1

# Creating of multiple objects with the same class
obj2 = myClass()
obj3 = myClass()
obj4 = myClass()

obj2.y
obj3.x
```

```python
# The _init_() method 
# the init() method is used to assign values to the object properties
class Person:
    def __init__(self, name, age, codeNo):
        self.name = name,
        self.age = age,
        self.codeNo = codeNo

p1 = Person('ram', 33, 8790)
print(p1.name)
```

```python
# Setting the default values for the parameters
class Fruit:
    def __init__(self, name, weight= '77g'):
        self.name = name,
        self.weight = weight

f1 = Fruit('apple')
print(f1.weight)
        
```

```python
# Some practice problems on Classes and objects
# Create  a class student with name marks  and add method to display name and marks

class Student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks
    def showMarks(self):
        print(f"{self.name} has obtained {self.marks}")

s1 = Student('hari', 66)

s1.showMarks()
```

```python
# Creating a class rectangle
class Rectangle:
    def __init__(self, length, breadth):
        self.length = length
        self.breadth = breadth
        
    def areaOfRec(self):
        print(f"Area of rectangle is {self.length * self.breadth} sq.")

# creating objec
obj = Rectangle(14, 10)
obj.areaOfRec()
```
