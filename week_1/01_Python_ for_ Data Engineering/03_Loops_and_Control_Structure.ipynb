```python
# lOOPS AND Control Structure in Python 

# Printing the number from 1 to 10 using the for loop and range in python 
for i in range(1,10):
    print(i)

# Printing only the even numbers from 1 to 20
for i in range (1, 20):
    if( i % 2 == 0):
        print(i)

# taking a number and printing its multiplication upto 10
number = int(input("Enter the number for the multiplication"))

for i in range(1, 10):
    print( f"{number} * {i} = {number * i}")
```

```python
# Conditional and Loop Combination 

# Print numbers from 1 to 20 

for i in range (1, 21):
    if( i % 3 == 0 and i % 5 == 0):
        print("Fizz Buzz")
    elif i % 3 == 0:
        print("Fizz")
    elif i % 5 == 0:
        print("Buzz")
    else:
        print(i)
```

```python
# Count how many numbers between 1 and 100 are divisible by 7
finalCount = 0
for i in range(1, 101):
    if( i % 7 == 0):
        finalCount += 1
print(finalCount)
```
