```python
# split() and join() in python 
sentence = 'Earth is the third planet in the solar system.'

words = sentence.split(' ')

newSentence = ' '
for word in words:
    word = word.upper()
    newSentence += word + " "
print(newSentence)
```

```python
# Dictionaries in Python
# Built in python data structure for mapping of keys and value

numbers = { 'one' : 1 , 
            'two' : 2, 
            'three' : 3
          }
# in this case 'one', 'two', 'three' are the keys where 1 , 2, 3 are the values

# you can access by the key
numbers['two']

# adding of another key and value in dicitionaries
numbers['ten'] = 10
numbers['eleven'] = 11

# changing the first value
numbers['one'] = 'pluto'
print(numbers)
```

```python
numbers

# looping through dictionaries

for num in numbers.values():
    print(num)
```

```python
for key, value in numbers.items():
    print(key, value)
```

```python
# Some practice problems on dictionaries

# Creating the dictionary for the car: brand, model and year
car = {
       'brand': 'Toyota',
        'model': 'Tacoma',
        'year' : 2017
}

print(car)
print(car['brand'])
print(car['year'])
print(car['model'])

# Exercise 2
person  = {"name" : "ram", 
           "age": 25
          }
person["city"] = 'Pokhara'
person['age']  = 26
print(person)


#Looping through the dicitionaries and printing
marks = { "math": 80, 
          "science": 75, 
          "english": 90
        }

for key, value in marks.items():
    print(f"{key} -> {value}")


# Count how many times each letter appears
word = "banana"
char_count = {}

for letter in word:
    if letter in char_count:
        char_count[letter] += 1
    else:
        char_count[letter] = 1
print(char_count)
```

```python
# Working with Nested Data structure list of dictionaries
students = [
    { "name" : "Ram", "marks": 80},
    {"name" : "Shyam", "marks" : 95}, 
    {"name": "Hari", "marks": 70}
]
# accessing the data

students[0]
students[-1]["name"]
students[-1]["marks"]

for student in students:
    print(student["name"], student["marks"])
```

```python
# Exercises to practice
students = [
    {"name": "Ram", "marks": 80},
    {"name": "Shyam", "marks": 95},
    {"name": "Hari", "marks": 70}
]

for student in students:
    print(f"{student["name"]} scored {student["marks"]} in the examination")

# Exercise 3
students = [
      {
        "name": "Ram",
       "marks":
               {
                "math": 80,
                "science": 70
               }
      },
      {
          "name": "Shyam", 
          "marks": {
              "math": 90, 
              "science": 85}
      }
]

for student in students:
   if student["name"] == 'Ram':
       print(f"{student['name']} scored {student['marks']['math']} in mathematics.")
   if student["name"] == "Shyam":
       print(f"{student['name']} scored {student['marks']['science']} in science.")


# Calculate the total marks of student

for student in students:
    if student["name"] == 'Ram':
        total = student['marks']['math'] + student['marks']['science']
        print(f"{student['name']} scored total marks of {total}")
    elif student["name"] == 'Shyam':
        total = student['marks']['math'] + student['marks']['science']
        print(f"{student['name']} scored total marks of {total}")
    
```
