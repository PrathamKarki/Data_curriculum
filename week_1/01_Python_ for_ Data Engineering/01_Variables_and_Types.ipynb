```python
# Variables and Types in Python

# Declaring of variable
# Python is a object oriented, but not a statically typed programming language unlike C, C++, typescript

name = 'pratham'
print(name)
```

```python
# Working with different types of variables in python 
# String
singleQuote = 'hello how are you?'

doubleQuote = "What's that?"

print(singleQuote)
print(doubleQuote)

# String concatenation 

greeting = 'hello'
fullname = 'pratham karki'

result = greeting + " " + fullname

print(result)
```

```python
# floats in python 

decimalValues = 1.34
secondValue = 4

print(decimalValues + secondValue)

anothervalue = float(13)
print(anothervalue)
```

```python
# assignment operator in python 
a = 4
b = 5

c, d = 30, 20
print(c,d)
```

```python
# Checking the dataypes with isinstance

string = 'hello'
number = 12
floatval = 12.76

if isinstance(string, str) and string == "hello":
    print("It is a string")
if isinstance(floatval, float) and number == 15:
    print("It is a float and it's value is", number)
else:
    print("It is a number and it's value is", number)
```
