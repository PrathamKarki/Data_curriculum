```python
# Lists in Python 
# List in python represents a sequence of value which can contain multiple datatypes
# creating a list
planets = ['Mercury', 'Venus', 'Earth', 'Mars', 'Jupiter', 'Saturn', 'Uranus', 'Neptune']

# accessing of the elements
planets[0] # accessing the first planet
planets[:4]
planets[:-2]

# nested list with different data types
information = [1, ['apple', 2, 'banana'], 3, 5 ,['4', None, 123.5]]

# accessing the informations
information[1][0]
information[-1][2]
```

```python
# Creating and acessing of list 

listOfNum = [5,7,8,12,15]
print(listOfNum[0])
print(listOfNum[-1])
print(listOfNum[2])

# Modifying of the elements in the list
nums = [10,20,30,40,50]
nums[1], nums[-1] = 25, 55

#Looping Through the list
fruits = ["apple", "banana", "mango"]

for fruit in fruits:
    print(fruit)
```

```python
# Sum of lists
nums = [1,2,3,4,5]

sum = 0;

for num in nums:
    sum += num
print(sum)
```

```python
# Count for even numbers in the list 
nombers = [1,2,3,4,5,6]
evenCount = 0;

for num in nombers:
    if num % 2 == 0:
        evenCount += 1
print(evenCount)
```
